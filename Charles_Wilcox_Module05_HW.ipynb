**Problem 1:** In the given data set below, identify the number of missing values by column; handle the missing values using SimpleImputer from scikit-learn library:

filling the missing values in col1 by most frequent number (hint: strategy='most_frequent').

````python
import pandas as pd
from io import StringIO

csv_data = \
'''
col1,col2
,2.0
,
10.0,11.0
1, 3 
1, 11
'''

df = pd.read_csv(StringIO(csv_data))
df
````

````python
#check for nulls by column name
df.isnull().sum()
````

````python
#impute missing values in column one using strategy='most_frequent'
from sklearn.impute import SimpleImputer
import numpy as np

col1 = pd.DataFrame(df['col1'])
imr = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imr = imr.fit(col1.values)
imputed_data = imr.transform(col1.values)

df['col1'] = imputed_data
df
````

**Problem 2:** Explain the transformer and estimator in scikit-learn. What are the differences between them?

Transformers are object that transform the dataset in on that is ready to be used for modeling. Examples of these transformations are scaling and one-hot encoding.
Estimators are object that learn from the data, such as classification, regression, clustering, etc.
Typically, any number of transformation are completed prior to using the estimator.

**Problem 3:** In the given data set below, standardize the numeric variables and generate the dummy variables for the categorical variables using one-hot encoding.

````python
import pandas as pd
import numpy as np
np.random.seed(123)

df = pd.DataFrame(np.random.rand(10, 2),
                 columns = ['col1', 'col2'])
df['col3'] = list(''.join(np.random.choice(list('abc')) for _ in range(len(df))))
df
````

````python
from sklearn.preprocessing import StandardScaler

#standardize numeric variables
stdsc = StandardScaler()
df[['col1','col2']] = stdsc.fit_transform(df[['col1','col2']])
df
````

````python
#generate dummy variables for the categorical variables
#when applied to a dataframe, get_dummies only converts string columns
pd.get_dummies(df[['col1', 'col2', 'col3']], drop_first=True)
````


**Problem 4:** Assume col3 is the target, separate the original data set in #3 into 80% training and 20% test.

**Problem 5:** Load scikit-learn's diabetes data set by load_diabetes(), pick the top 3 most discriminative features using Random Forest.

````python
from sklearn.ensemble import RandomForestClassifier

#load diabetes dataset
from sklearn import datasets
diabetes = datasets.load_diabetes()
X = diabetes.data
y = diabetes.target

#random forest
forest = RandomForestClassifier(n_estimators=500,
                                random_state=1, max_features=0.9)
forest.fit(X, y)
importances = forest.feature_importances_
feature_labels = diabetes.feature_names

#print importances and feature lables
print(importances)
print(feature_labels)
````

````python
#rank importance in decending order
indices = np.argsort(importances)[::-1]

for f in range(X.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, 
                            feature_labels[indices[f]], 
                            importances[indices[f]]))
````

The top 3 most discriminative feature are bmi, s5, and s6.
