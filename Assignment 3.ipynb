```python
#import pandas, import csv file
import pandas as pd
home_games = pd.read_csv\
('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/HomeGames.csv')
```

```python
#first, check the data types of the dataset
home_games.dtypes
```

```python
#converted the span.last column into a datetime dtype
last_home_game = pd.to_datetime(home_games['span.last'], format='%Y-%m-%d')

last_home_game
```

```python
#created a new column named "last_hg" for the last home game
home_games['last_hg'] = last_home_game

home_games
```

```python
# used .max() to find the maximum date in the first_hg column
#named the date max_fhg
max_lhg = home_games.last_hg.max()

print(max_lhg)
```

```python
#subtracted the maximum date max_lhg from the column last_hg
#saved in a new column called days_since
home_games['days_since'] = max_lhg - home_games.last_hg

home_games
```

```python
#changed days_since to months by using .astype('timedelta64[M]')
#added months data to data set in column named months_since
home_games['months_since'] = home_games['days_since'].astype('timedelta64[M]')

home_games
```

```python
#saved dataframe as a csv file using .to_csv
#used index=False so there is no index value column 
home_games.to_csv('home_games_clean.csv', index=False)
```
