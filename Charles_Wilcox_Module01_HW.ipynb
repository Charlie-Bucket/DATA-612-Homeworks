Problem 3. 

Select the rows of men's gold medals in singles badminton, and sort the results alphabetically by the player's names.

```python
#selecting rows that include God, Men, Singles, & Badminton
gm_men_sing_bad = df[(df.Medal == "Gold") & (df.Gender == "Men") & (df.Event == "singles") & (df.Discipline == "Badminton")]
#sorting those rows by name (column 'Athlete')
badminton_sort = gm_men_sing_bad.sort_values(by= "Athlete")
badminton_sort.head()
```

Problem 4.

List the top three countries (NOC column) that won the most medals from 1984 (inclusive) to present.

```python
#Selecting rows where the year (Edition) is greater than or equal to 1984
recent_medals = df[df.Edition >= 1984]
#counting the instances the country code shows up in the recent_medals dataframe
medal_counts = recent_medals.NOC.value_counts()
#returning the top 3 
medal_counts.head(3)
```

Problem 5.

Plot the number of medals of the top 3 athlete in 2008 using matplotlib.pyplot.

- use bar chart

- use value_counts(); note the returned results is a Series. So, the x axis is the index of Series.

```python
#Selecting only rows where Edition = 2008
year_2008 = df[df.Edition == 2008]

#creating a series of the value counts for each Athlete
counts_2008 = df.Athlete.value_counts()

#reducing the series to the top 3 Athletes
top_3 = counts_2008.head(3)

#ploting the top 3 Athletes in a bar chart
import matplotlib.pyplot as plt
plt.bar(top_3.index, top_3)
```

Problem 6.

Use list comprehension to define a new column "is_US". When the country is USA, the value of is_US is True, otherwise the value is False.

```python
#create a new column named is_US and use list comprehension to fill in boolean values
df['is_US'] = [True if x =='USA' else False for x in df['NOC']]
#view sample
df.sample(10)
```

Problem 7.

Hide the medal information for 1988 olympics; i.e., when edition is 1988, change the Medal as "Unknown".

```python
#update the subset of data where Edition = 1988, changing Medal to Unknown
df.loc[df['Edition'] == 1988, 'Medal'] = 'Unknown'

#view dataset to review change
year_1988 = df[df.Edition == 1988]
year_1988.head()
```
