```python
import pandas as pd
fruit = pd.read_csv("https://raw.githubusercontent.com/franklin-univ-data-science/data/master/fruit.csv")
fruit
```

**2. Define the features as mass and height, and the target as the variable "isLemon", which is derived by yourself (if fruit is lemon then 1 else -1). Print the feature matrix and target vector, respectively.**

```python
import numpy as np

#create a target vector named "isLemon"
y = fruit.iloc[0:, 0].values
isLemon = np.where(y == "lemon", 1, -1)

#create a feature matrix from mass and height
X = fruit.iloc[0:, [1,2]].values

print("Feature Matrix:")
print(X)
print()
print("Target Vector:")
print(isLemon)
```

**3. Draw a scatter plot of the binary classification (isLemon) by two features (mass, height)**

**4. Add x0 vector to the feature matrix X. Apply the perceptron algorithm to predict the classification (isLemon). What is the smallest number of misclassified samples you can get? (hint: try a very large number of epoch because the optimization of our current perceptron method is not efficient)**

```python
#create a vector of ones the length of the isLemon vector
m = y.size
vector_ones = np.ones_like(isLemon).reshape(m, 1)

#join vector_ones to the feature matrix
X1 = np.hstack([vector_ones, X])
X1
```

```python
class Perceptron():
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.

    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    errors_ : list
      Number of misclassifications (updates) in each epoch.

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
          Training vectors, where n_samples is the number of samples and
          n_features is the number of features.
        y : array-like, shape = [n_samples]
          Target values.

        Returns
        -------
        self : object

        """
        #initialize the weights
        rgen = np.random.RandomState(self.random_state)      #rgen is a NumPy random number generator seeded with a user-specified random seed so that we can reproduce previous results
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1]) # get the initial weight
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_ += update * xi
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return X @ self.w_
  

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
#train the perceptron algorithm
ppn = Perceptron(eta=0.1, n_iter=100000, random_state=1)
ppn.fit(X1, isLemon)

#plot the misclassification error for each epoch
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)

plt.plot(range(1, ppn.n_iter + 1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')

plt.show()
```

At 80,000 epochs, the number of misclassified samples drops to 0.

```python
#check misclassified samples
y_pred = ppn.predict(X1)
print('Misclassified samples: %d' % (isLemon != y_pred).sum())
```
